```python
# This is the function to convert the ordinal columns from the string values to numerical values. Done with lots help from Alanna!

ordinal_change = {"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}
def convert_cond_qual(df):
    for col in df.loc[:, df.dtypes == 'O']:
        df[col] = df[col].replace(ordinal_change)
    return df

    
    
    
```
