```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../datasets/train.csv")
test = pd.read_csv("../datasets/test.csv")
```

```python
#Create a heatmap to show correlation between all of the features (from Visualization lesson)
corr = train.corr()
plt.figure(figsize=(20, 20))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, mask=mask);
```

```python
# Create a features list with the features that most positively correlate with Sale Price
features = ["Lot Frontage", "Lot Area", "BsmtFin SF 2", "Overall Qual", "Year Built", "Year Remod/Add", "Mas Vnr Area", 
            "Total Bsmt SF", "1st Flr SF", "Gr Liv Area", "Full Bath", "TotRms AbvGrd", 
            "Garage Yr Blt", "Garage Cars", "Garage Area", "Wood Deck SF", "Open Porch SF"]
```

```python
X = train[features]
y = train["SalePrice"]

print('X: ', X.shape)
print('y: ', y.shape)
```

```python
# From Kaggle lesson on Missing Values: https://www.kaggle.com/jillanisofttech/exercise-missing-values
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Define a function that gives the mean absolute error for a train-val model
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=10, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score

# Split the training data into training and validation sets
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```

```python
# From Kaggle lesson on Missing Values: https://www.kaggle.com/jillanisofttech/exercise-missing-values

from sklearn.impute import SimpleImputer

# Impute the mean value of a column into the rows with missing values 
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(train_X))
imputed_X_valid = pd.DataFrame(my_imputer.transform(val_X))

# Imputation removed column names; put them back
imputed_X_train.columns = train_X.columns
imputed_X_valid.columns = val_X.columns

print(f"The mae for this train/test model is: {score_dataset(imputed_X_train, imputed_X_valid, train_y, val_y)}")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(imputed_X_train, train_y)
```

```python
model.score(imputed_X_train, train_y)
```

```python
model.score(imputed_X_valid, val_y)
```

```python
cross_val_score(model, imputed_X_train, train_y).mean()
```

```python
# Testing the full training set
X_filled = pd.DataFrame(my_imputer.fit_transform(X))
X_filled.columns = X.columns
X_filled.isna().sum()
```

```python
home_model = LinearRegression()
home_model.fit(X_filled, y)
predict = home_model.predict(X_filled)
```

```python
from sklearn import metrics
model_mse = metrics.mean_squared_error(y, predict)
model_mse
```
