```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importimg the train dataset from the csv created in the  "notebook for cleaning" 
train = pd.read_csv("../datasets/train_cleaned.csv")
test = pd.read_csv("../datasets/test.csv")
```

```python
train.head()
```

```python
#Create a heatmap to show correlation between all of the features (from Visualization lesson)
corr = train.corr()
plt.figure(figsize=(20, 20))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, mask=mask);
```

```python
# Create a features list with the features that most positively correlate with Sale Price (> .60)
features = ["overall_qual", "total_bsmt_sf", "1st_flr_sf", "garage_cars", "garage_area"]
# "Lot Frontage", "Lot Area", "BsmtFin SF 2", "Year Built", "Year Remod/Add", "Gr Liv Area", "Full Bath", "TotRms AbvGrd", "Garage Yr Blt",  "Open Porch SF" 
```

```python
# Checking to see if any variable breaks the L assumption
sns.pairplot(data=train, x_vars=["overall_qual", "total_bsmt_sf", "1st_flr_sf", "garage_cars", "garage_area"], 
             y_vars = ["saleprice"], palette="coolwarm");
```

```python
X = train[features]
y = train["saleprice"]

print('X: ', X.shape)
print('y: ', y.shape)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score

# Split the training data into training and validation sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
cross_val_score(model, X_train, y_train).mean()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
cross_val_score(model, X_train, y_train).mean()
```

```python
preds = model.predict(X_test)
```

```python
residuals = y_test - preds
```

```python
plt.figure(figsize=(10,10))
plt.hist(preds, color="green", bins=20)
plt.xlim(0, 500_000)
# plt.yscale("log")
plt.axvline(y.mean(), color="purple")
plt.show();
```
