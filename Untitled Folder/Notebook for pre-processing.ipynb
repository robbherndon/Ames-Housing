```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../datasets/train_cleaned.csv")
test = pd.read_csv("../datasets/test_cleaned.csv")
```

```python
train.drop(columns = "Unnamed: 0", inplace = True)
```

```python
test.drop(columns = "Unnamed: 0", inplace=True)
```

```python
train["bsmt_cond"].value_counts()
```

```python
train["overall_combined"] = train["overall_cond"] * train["overall_qual"]
```

```python
test["overall_combined"] = test["overall_cond"] * test["overall_qual"]
```

```python
train["overall_combined"].corr(train["saleprice"])
```

```python
train["garage_overall"] = train["garage_cond"] * train["garage_qual"]
```

```python
test["garage_overall"] = test["garage_cond"] * test["garage_qual"]
```

```python
train["garage_overall"].corr(train["saleprice"])
```

```python
train["bsmt_overall"] = train["bsmt_cond"] * train["bsmt_qual"]
```

```python
test["bsmt_overall"] = test["bsmt_cond"] * test["bsmt_qual"]
```

```python
train["bsmt_overall"].corr(train["saleprice"])
```

```python
train["total_qual"] = train["garage_qual"] * train["bsmt_qual"] * train["kitchen_qual"] * train["exter_qual"]
```

```python
test["total_qual"] = test["garage_qual"] * test["bsmt_qual"] * test["kitchen_qual"] * test["exter_qual"]
```

```python
train["total_qual"].corr(train["saleprice"])
```

```python
corr = train.corr()
plt.figure(figsize=(30, 20))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, mask=mask);
```

```python
train.to_csv("../datasets/train_feat_eng.csv")
```

```python
test.to_csv("../datasets/test_feat_eng.csv")
```
