# Create Polynomial Features

```python
features = ["id","lot_frontage", "lot_area", "year_built", "year_remod/add", 
            "total_bsmt_sf", "total_qual", "1st_flr_sf", "gr_liv_area", "bsmt_full_bath",
            "full_bath", "half_bath", "totrms_abvgrd", "fireplaces", "garage_yr_blt", "garage_cars", 
            "garage_area", "wood_deck_sf", "open_porch_sf"]
```

```python
X_train = train[features]
```

```python
X_test = test[features]
```

```python
#Instantiate for train
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
poly = PolynomialFeatures(include_bias=False)
# Create X_poly for train
X_poly_train = poly.fit_transform(X_train)
```

```python
#Instantiate for test
X_poly_test = poly.transform(X_test)
```

```python
# Create X_poly for train
# X_poly = poly.transform(X_train)
```

```python
# Create X_poly for test
# X_poly_test = poly.transform(X_test)
```

```python
# View X_poly in a DataFrame
poly.get_feature_names(features)
```

```python
# Create poly_train Dataframe
poly_train = pd.DataFrame(X_poly_train, columns=poly.get_feature_names(features))
poly_train
```

```python
# Create poly_test Dataframe
poly_test = pd.DataFrame(X_poly_test, columns=poly.get_feature_names(features))
poly_test
```

```python
# Merge original train DF with poly_train to get new DF
train_and_poly = pd.merge(train, poly_train, on="id")
train_and_poly
```

```python
# Merge original test DF with poly_test DF to get new DF
test_and_poly = pd.merge(test, poly_test, on="id")
test_and_poly
```
