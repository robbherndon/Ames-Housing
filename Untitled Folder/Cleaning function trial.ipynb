```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../datasets/train.csv")
test = pd.read_csv("../datasets/test.csv")
pd.set_option('display.max_columns', 81)
```

```python
train.columns = [col.replace(" ", "_").lower() for col in train.columns]
```

```python
test.columns = [col.replace(" ", "_").lower() for col in test.columns]
```

```python
categorical_cols = [col for col in train.columns if train[col].dtype == "object"]
numerical_cols = [col for col in train.columns if train[col].dtype == "float64" or "int64"]
```

```python
ordinal_change = {"NA":0, "Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}
def convert_cond_qual(df):
    for col in df.loc[:, df.dtypes == 'O']:
        df[col] = df[col].replace(ordinal_change)
    return df
```

```python
convert_cond_qual(train)
```

```python
convert_cond_qual(test)
```

```python
def cat_convert_nan(df):
    for col in df.loc[:, df.dtypes == 'O']:
        df[col] = df[col].fillna("NaN")
    for cols in df.loc[:, df.dtypes == 'int64']:
        df[cols] = df[cols].fillna(df[cols].median())
    for columm in df.loc[:, df.dtypes == 'float64']:
        df[columm] = df[columm].fillna(df[columm].median())
    return df
```

```python
cat_convert_nan(train)
```

```python
cat_convert_nan(test)
```
